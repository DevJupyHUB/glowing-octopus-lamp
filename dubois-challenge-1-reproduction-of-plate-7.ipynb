```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
```

```python
df = pd.read_csv('/kaggle/input/dubois-data-portrait-challenge/challenge-only/challenge01/data.csv')
```

```python
df.info()
```

```python
# Create figure and axis
fig, ax = plt.subplots(figsize = (8.15, 10.24), facecolor = '#e4d1b4')

# Set axis limits and labels
ax.set_xlim([100, 0])
ax.set_ylim([1790,1890])
ax.set_xticks(list(range(100,-5,-5)))
ax.set_yticks(list(range(1790,1900,10)))

# Adjust Axes to occupy 60% width 
ax.set_position([0.22, 0.15, 0.60, 0.7])

# Plot the first line
plt.plot(df['Colored'], df['Year'], color = 'black', alpha = 0.8, lw = 1)

# Plot the second line
plt.plot(df['White'], df['Year'], color = 'black', ls = ('dashed'), alpha = 0.8, lw = 1)

# Set title
ax.set_title('COMPARATIVE INCREASE OF WHITE AND COLORED\n POPULATION OF GEORGIA.', fontsize = 18, pad = 35)

# Set x label
ax.set_xlabel('PERCENTS', fontsize = 8,alpha=0.6, labelpad = 20)

# Add grids
ax.grid(True, color = '#eeb59e')

# Apply alpha for the grid frame 
for spine in ax.spines.values():
    spine.set_alpha(0.4)  

# Change x tick label size and remove ticks
ax.tick_params(axis = 'x', which = 'major', labelsize = 8)  
ax.tick_params(axis = 'both', which = 'both', width = 0)  

# Apply alpha to x-tick labels
for tick in ax.get_xticklabels():
    tick.set_alpha(0.6)  
    
# Apply alpha to y-tick labels
for tick in ax.get_yticklabels():
    tick.set_alpha(0.6)  

# Add legend
legend = ax.legend(['= COLORED', '= WHITE'], loc = 'lower center', fontsize = 8,
                   bbox_to_anchor = (.5,-.10), frameon = False, ncol = 2, columnspacing = 30)

# Apply alpha for legend labels 
for text in legend.get_texts():
    text.set_alpha(0.6)  

# Set fig facecolor
ax.set_facecolor('#e4d1b4')

# Save the fig
plt.savefig("plate07.jpg")
```

```python
# Load the images
image1 = Image.open("/kaggle/input/dubois-data-portrait-challenge/challenge-only/challenge01/original-plate-07.jpg")
image2 = Image.open("plate07.jpg")

# Convert the image to a numpy array 
image_np1 = np.array(image1)
image_np2 = np.array(image2)

# Create figure with two subplots 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (8.15, 10.24))

# Left subplot
ax1.imshow(image_np1)
ax1.axis('off')  # Turn off axis for the image
ax1.set_title('Original')

# Right subplot
ax2.imshow(image_np2)
ax2.axis('off')  # Turn off axis for the image
ax2.set_title('Reproduction')

# Adjust the layout 
plt.tight_layout()

# Show the figure
plt.show()
```
